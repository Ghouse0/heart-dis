# Encoding Categorical Features

```python
categorical_cols=['chest_pain','thalassemia','st_slope']

for col in categorical_cols:
    df[col]=df[col].astype(int)
    
    
df_encode=pd.get_dummies(df,columns=categorical_cols,prefix_sep='_',dtype=int)

df_encode.dtypes
```

```python
x=df.drop('diagnosis',axis=1)

y=df['diagnosis']
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler= MinMaxScaler()
x_scaled =scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=4)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

lr_model=LogisticRegression()
lr_model.fit(x_train,y_train)
lr_cv_result=cross_val_score(lr_model,x_train,y_train,cv=10)
lr_mean_score=round(lr_cv_result.mean(),4)


print("mean score from cross_validation: ",lr_mean_score)


```

# Linear Discriminant Analysis (LDA)

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

ldr_model=LinearDiscriminantAnalysis()
ldr_model.fit(x_train,y_train)
ldr_cv_result=cross_val_score(ldr_model,x_train,y_train,cv=10)
ldr_mean_score=round(ldr_cv_result.mean(),4)

print("mean score from cross_validatiom:",ldr_mean_score)
```

# K-Nearest Neighbors (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model=KNeighborsClassifier()
knn_model.fit(x_train,y_train)
knn_cv_results=cross_val_score(knn_model,x_train,y_train,cv=10)
knn_mean_score=round(knn_cv_results.mean(),4)
knn_std_score=round(knn_cv_results.std(),4)

print('mean score from cross_validation:',knn_mean_score)
print("standard devation of the score:",knn_std_score)
 
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

dt_model=DecisionTreeClassifier()
dt_model.fit(x_train,y_train)
dt_cv_result = cross_val_score(dt_model,x_train,y_train,cv=10)
dt_mean_score=round(dt_cv_result.mean(),4)

print("mean score from cross_validation:",dt_mean_score)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

num_trees=100
max_feature='sqrt'

rf_model=RandomForestClassifier(n_estimators=num_trees,max_features=max_feature)
rf_model.fit(x_train,y_train)
rf_cv_result =cross_val_score(rf_model,x_train,y_train,cv=10)
rf_mean_score=round(rf_cv_result.mean(),4)

print("mean score from cross-validation:",rf_mean_score)
```
